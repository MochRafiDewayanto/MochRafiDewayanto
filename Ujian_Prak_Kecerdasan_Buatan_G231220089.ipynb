```python
# Memanggil Library
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Masukkan dataset dengan file yang sudah dibuat
dataset = pd.read_excel('datauasAI.xlsx')
dataset.head(10)
```

```python
# Membuat atribut regular
x = dataset.drop(['Streaming'], axis=1)
x.head()
```

```python
# Membuat atribut target/class
y = dataset['Streaming']
y.head()
```

```python
#import gausian naive bayes model
from sklearn.naive_bayes import GaussianNB
```

```python
#memanggil fungsi Gaussian Naive Bayes
nbc = GaussianNB()
```

```python
#Memasukkan data training ke dalam fungsi klassifikasi naive bayes
data_training = nbc.fit(x,y)
data_training
```

```python
# melakukan prediksi pada data training
y_predict = data_training.predict(x)
print(y_predict)
```

```python
# melakukan prediksi pada data input --> data yang akan diprediksi
Surat = 1 # 1 = Ya
Sosmed = 1 # 1 = Ya
Intenert = 0 # 0 = Tidak
JenisKelamin = 0 # 0 = P
data_testing = [[Surat, Sosmed, Intenert, JenisKelamin]]
data_testing
```

```python
#Melakukan prediksi berdasarkan data input
y_pred = data_training.predict(data_testing)
print(y_pred)
```

```python
#Mencetak hasil prediksi promosi
# hasil 0 = Terlambat, 1 = tepat
if y_pred ==0:
  hasil = 'Tidak'
elif y_pred == 1:
  hasil = 'Ya'
else :
  hasil = 'Error'
print('Hasil Prediksi Promosi : ', hasil)
```

```python
#menghitung akurasi --> melihat skor akurasi
from sklearn.metrics import accuracy_score
print('Nilai akurasi = %0.2f' % accuracy_score(y,y_predict))
```

```python
# Mencetak laporan hasil klasifikasi
from sklearn.metrics import classification_report
print(classification_report(y,y_predict))
```
